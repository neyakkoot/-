2. தரவுக் கோப்பை உள்ளீடு செய்து ஆய்வு செய்தல்

1) தரவுக் கோப்பு எடுத்தல்

2) அதனை உள்ளீடு செய்வதற்குரிய முறைமையைத் தெரிவுசெய்தல்

    2.1) உள்ளீடாகக் கேட்டல் (input) அல்லது

    2.2) தானாகவே தேடி எடுத்து ஆய்வு செய்தல் (open, withopen)

3) உள்ளீடு செய்த கோப்பினைப் படித்தல் (read, readlines)

4) தரவுப் பகுப்பாய்வு செய்தல் (split)

5) விளைவுக்கோப்பு உருவாக்கல் (write)

```python
# உருவாக்குநர் - தகவலுழவன்
# பயனாகும் பைத்தான் குறிச்சொற்கள்
   ## import, input, open, 'r', .readlines(), for, in, len, if , !=,
   ## .split(), .replace('(','').replace(')',''),
   ## .isdigit(), if not, print, time.sleep(0),
   ## with open, 'a', as, write, '\n'

import time, re

உள்ளீட்டுக்கோப்பு = input("உள்ளிடும் கோப்பின் பெயரைத் தருக : ")
மூலக்கோப்பு = open(உள்ளீட்டுக்கோப்பு, 'r')
பத்திக்கோப்பு = மூலக்கோப்பு.readlines()

for பத்தி in பத்திக்கோப்பு:
	பத்தியெண்ணிக்கை = len(பத்தி)
	if பத்தியெண்ணிக்கை !=1:
		வரிப்பட்டி = பத்தி.split()
		for வரி in வரிப்பட்டி:
			வரிசுத்தம் = வரி.replace('(','').replace(')','')
			வரிசுத்தம்எண் = வரிசுத்தம்.isdigit()
			if not வரிசுத்தம்எண்:
				print(வரிசுத்தம்)
				#time.sleep(0)
				with open('கடிதங்கள்_சொற்பிரிப்பு.text','a') as resultFile:
					resultFile.write(வரிசுத்தம் + '\n')
```

```python
# உருவாக்குநர் - தகவலுழவன்
#import time

பயனாகும்கோப்பு1 = input("உள்ளிடும் கோப்பின் பெயரை உள்ளிடவும் : ")
மூலக்கோப்பு = open(பயனாகும்கோப்பு1, 'r')
மூலக்கோப்புத்தரவுப்பட்டி = மூலக்கோப்பு.readlines()

for வரி in மூலக்கோப்புத்தரவுப்பட்டி:
    வரிமுன்பின்வெற்றிடம் = வரி.strip()
    வரியெழுத்தெண்ணிக்கை = len(வரிமுன்பின்வெற்றிடம்)

    if not வரியெழுத்தெண்ணிக்கை == 0:
        if 'பாடல்' in வரி:
            பாடலெண் = வரி.split('-')[-1]
            பாடலெண்சரம் = 'பாடல்' +  str(பாடலெண்).strip() + '-'
            வரியெண் = 0

        if not 'பாடல்' in வரி:
            வரியெண் = வரியெண்+1
            வரியெண்வடிவம் = 'வரி' + str(வரியெண்) + '- '
            வரியெண்ணுடன்வரி = வரியெண்வடிவம் + வரிமுன்பின்வெற்றிடம்
            பாடலெண்வரியெண்வரி = பாடலெண்சரம் + வரியெண்ணுடன்வரி

            வரிசொற்கள்பிரி = பாடலெண்வரியெண்வரி.split()
            பாடலெண்வரியெண் = வரிசொற்கள்பிரி[0]
            பாடலெண்வரியெண்மீதி = வரிசொற்கள்பிரி[1:]

            சொல்லெண் = 0
            for பாடற்சொல் in பாடலெண்வரியெண்மீதி:
                சொல்லெண் = சொல்லெண்+1
                சொல்லெண்சரம் = 'சொல்' + str(சொல்லெண்)

                இறுதிவடிவம் = பாடற்சொல் + ' (' + பாடலெண்வரியெண்.replace('பாடல்','').replace('-வரி',':') + சொல்லெண்சரம்.replace('சொல்','') + ')' #.replace('','')
                print(இறுதிவடிவம்)

                with open('0testSeparated.text','a') as resultFile:
                    resultFile.write(இறுதிவடிவம்+'\n')
```
